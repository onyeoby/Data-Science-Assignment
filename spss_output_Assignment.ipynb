```python
file["TSE_NEW"] = file[["TSE1","TSE2","TSE3","TSE4","TSE5"]].mean(axis = 1)
```

```python
file["EII_NEW"] = file[["EII1","EII2","EII3","EII4","EII5","EII6","EII7","EII8"]].mean(axis = 1)
```

```python
file["SELS_NEW"] = file[["SELS1","SELS2","SELS3","SELS4","SELS5"]].mean(axis =1)
```

```python
file["OTHER_FACTORS"] = file[["Entertainment","Shopping","HealthandFitness","Education","Travel","SocialNetworking","Navigation","Productivity","Finance","Business"]].sum(axis = 1)
```

```python
y = file1["ATT_NEW"]
x = file1[["Income","TSE_NEW","EII_NEW","SELS_NEW","OTHER_FACTORS"]]
y = np.array(y)
x = np.array(x)
x = sm.add_constant(x)
model = sm.OLS(y,x)
fit = model.fit()
result = fit.summary()
print(result)
```
