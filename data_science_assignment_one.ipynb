```python
import numpy as np
import pandas as pd
```

```python
# Preparing and Cleaning the data
file = "NGA_relative_wealth_index.csv"
file = pd.read_csv(file)
```

```python
file
```

```python
# putting it in Data frame to enable us manipulate it
file = pd.DataFrame(file)
```

```python
file
```

```python
# To run the descriptive Analysis
file.describe()
```

```python
file.corr()
```

```python
# Running a simple Regression Analysis
import statsmodels.api as sm
```

```python
y = file["rwi"]
x = file["latitude"]
y = np.array(y)
x = np.array(x)
x = sm.add_constant(x)
model = sm.OLS(y,x)
fit = model.fit()
result=fit.summary()
print(result)
```

```python
y = file["rwi"]
x = file[["latitude","longitude"]]
y = np.array(y)
x = np.array(x)
x = sm.add_constant(x)
model = sm.OLS(y,x)
fit = model.fit()
result = fit.summary()
print(result)
```
